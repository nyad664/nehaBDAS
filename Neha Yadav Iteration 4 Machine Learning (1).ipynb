```python
import findspark
findspark.init('/home/ubuntu/spark-2.1.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('basics').getOrCreate()
```

```python
df = spark.read.csv('Datasets/Kaagle_Upload.csv',header=True,inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
#selecting
```

```python

from pyspark.sql import *
spark = SparkSession.builder.appName('logistic_regression_adv').getOrCreate()
```

```python
# Import the relevant packages
```

```python
from pyspark.ml.feature import (VectorAssembler,VectorIndexer,OneHotEncoder,StringIndexer)
```

```python
ur_indexer = StringIndexer(inputCol='urban_or_rural_area',outputCol='urbanrural')
weather_indexer = StringIndexer(inputCol='weather_conditions',outputCol='weather')
driversex_indexer = StringIndexer(inputCol='sex_of_driver',outputCol='driversex')
rsc_indexer = StringIndexer(inputCol='road_surface_conditions',outputCol='rsc')
sl_indexer = StringIndexer(inputCol='speed_limit',outputCol='sl')
accidentseverity_indexer = StringIndexer(inputCol='accident_severity',outputCol='label')
```

```python
urru_encoder = OneHotEncoder(inputCol='urbanrural',outputCol='urbanruralVec')
weather_encoder = OneHotEncoder(inputCol='weather',outputCol='weatherVec')
driversex_encoder = OneHotEncoder(inputCol='driversex',outputCol='driversexVec')
rsc_encoder = StringIndexer(inputCol='rsc',outputCol='rscVec')
sl_encoder = StringIndexer(inputCol='sl',outputCol='slVec')
label_encoder = StringIndexer(inputCol='label',outputCol='label')

```

```python
assembler = VectorAssembler(inputCols=['urbanruralVec','weatherVec','driversexVec',
                                       'rscVec','slVec'], outputCol="features")
```

```python
#pipeline
```

```python
from pyspark.ml import Pipeline
```

```python
pipeline = Pipeline(stages=[ur_indexer,weather_indexer,driversex_indexer,rsc_indexer,sl_indexer,accidentseverity_indexer,
                            urru_encoder,weather_encoder,driversex_encoder,rsc_encoder,sl_encoder,assembler])

```

```python
pipeline_model = pipeline.fit(df)
```

```python
pipe_df = pipeline_model.transform(df)
```

```python
pipe_df = pipe_df.select('label','features')
```

```python
pipe_df.describe()
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
train_data, test_data = pipe_df.randomSplit([0.7,0.3])
print("Training Dataset Count: " + str(train_data.count()))
print("Test Dataset Count: " + str(test_data.count()))
```

```python
# Random Forest  ITERATON 3
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction",metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test set accuracy = " + str(accuracy))
```

```python
from pyspark.ml.classification import LogisticRegression, OneVsRest
from pyspark.ml.evaluation import MulticlassClassificationEvaluator


# generate the train/test split.
(train_data, test_data) = pipe_df.randomSplit([0.7, 0.3])

# instantiate the base classifier.
lr = LogisticRegression(maxIter=5, tol=1E-6, fitIntercept=True)

# instantiate the One Vs Rest Classifier.
ovr = OneVsRest(classifier=lr)

# train the multiclass model.
ovrModel = ovr.fit(train_data)

# score the model on test data.
predictions = ovrModel.transform(test_data)

# obtain evaluator.
evaluator = MulticlassClassificationEvaluator(metricName="accuracy")

# compute the classification error on test data.
accuracy = evaluator.evaluate(predictions)
print("Test accuracy = %g" % (1.0 - accuracy))
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
print("Histograms below: ")

severity = df.groupBy('accident_severity').count().collect() 
categories = [i[0] for i in severity]
counts = [i[1] for i in severity]

ind = np.array(range(len(categories)))
width = 0.35
plt.bar(ind, counts, width=width, color='y')

plt.ylabel('counts')
plt.title('Response distribution')
plt.xticks(ind + width/2., categories)
```
